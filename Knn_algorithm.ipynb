```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('diabetes_patients.csv')

# Display the missing values count per column
print("Missing values per column before imputation:")
print(df.isnull().sum())

# Initialize the KNN imputer with n_neighbors = 5 (default)
imputer = KNNImputer(n_neighbors=5)

# Select the columns to be imputed (excluding the 'Disease' column if it's categorical)
columns_to_impute = ['plasma_CA19_9', 'LYVE1', 'CYFRA 21-1', 'Creatinine']
df_imputed = df.copy()
df_imputed[columns_to_impute] = imputer.fit_transform(df[columns_to_impute])

# Round the imputed values to 3 decimal places
df_imputed[columns_to_impute] = df_imputed[columns_to_impute].round(3)

# Display the missing values count per column after imputation
print("\nMissing values per column after imputation:")
print(df_imputed.isnull().sum())

# Save the imputed DataFrame to a new CSV file
df_imputed.to_csv('imputed_patient_data.csv', index=False)

print("\nImputation completed and saved to 'imputed_patient_data.csv'.")
```

```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.metrics import mean_squared_error, mean_absolute_error
import random

# Read the original and imputed CSV files into DataFrames
df_original = pd.read_csv('original_data.csv')
df_imputed = pd.read_csv('imputed_patient_data.csv')

# Introduce missing values randomly in a copy of the DataFrame to simulate missing data
np.random.seed(42)  # For reproducibility
missing_rate = 0.1  # 10% missing values

df_with_missing = df_original.copy()

# Calculate metrics to compare original and imputed data
metrics = {}
for col in ['plasma_CA19_9', 'LYVE1', 'CYFRA 21-1', 'Creatinine']:
    # Extracting original, imputed values, and mask
    original_values = df_original[col].values
    imputed_values = df_imputed[col].values
    mask = np.isnan(df_with_missing[col].values)
    
    # Remove NaN entries for comparison
    valid_original_values = original_values[~mask]
    valid_imputed_values = imputed_values[~mask]

    # Calculate Mean Squared Error (MSE) and Mean Absolute Error (MAE)
    mse = mean_squared_error(valid_original_values, valid_imputed_values)
    mae = mean_absolute_error(valid_original_values, valid_imputed_values)
    
    # Accuracy as the percentage of correctly imputed values
    threshold = 0.001  # Define a threshold for what is considered "correct"
    correct_imputations = np.abs(valid_original_values - valid_imputed_values) < threshold
    accuracy_percentage = np.mean(correct_imputations) * 100

    metrics[col] = {
        'Mean Squared Error (MSE)': mse,
        'Mean Absolute Error (MAE)': mae,
        'Accuracy (%)': accuracy_percentage
    }

# Display metrics for each column
for col, metrics_col in metrics.items():
    print(f"\nMetrics for {col}:")
    for metric_name, metric_value in metrics_col.items():
        print(f"{metric_name}: {metric_value:.2f}")

# Save the imputed DataFrame to a new CSV file
df_imputed.to_csv('imputed_patient_data.csv', index=False)
print("\nImputed DataFrame saved to 'imputed_patient_data.csv'.")
```
